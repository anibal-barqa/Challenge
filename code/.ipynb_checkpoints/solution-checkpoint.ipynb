### 1.4 - SYNTHETIC FEATURES <a class="anchor" id="synt"></a>

```python
# Temporada alta
df['temporada_alta'] = [1 if (x < dt.datetime(2017,3,4)) or (x >= dt.datetime(2017,12,15)) or ((x >= dt.datetime(2017,7,15)) and (x < dt.datetime(2017,8,1))) or ((x >= dt.datetime(2017,9,11)) and (x < dt.datetime(2017,10,1))) else 0 for x in df['Fecha-I']]
```

```python
# Diferencia en minutos
df['dif_min'] = (df['Fecha-O']-df['Fecha-I'])/pd.Timedelta(minutes=1)
```

```python
# Atraso de 15 minutos
df['atraso_15'] = [1 if x>15 else 0 for x in df['dif_min']]
```

```python
# Período del día
df['periodo_dia'] = ['mañana' if (x.time() >= dt.time(5,0)) and (x.time() < dt.time(12,0)) else 'tarde' if (x.time() >= dt.time(12,0)) and (x.time() < dt.time(19,0)) else 'noche' for x in df['Fecha-I']]
```

```python
# Cambio de destino
df['dif_destino'] = (df['Des-I']!=df['Des-O'])*1
```

```python
# Exportación de datos
df[['temporada_alta','dif_min','atraso_15','periodo_dia','dif_destino']].to_csv(r'..\data\synthetic_features.csv', index=False)
```

### 1.6 - ANÁLISIS DE AUTOCORRELACIÓN <a class="anchor" id="corr"></a>

##### Se realizará un análisis de autocorrelación, con el fin de excluir las variables con correlación mayor a 0.95.

```python
# Filtro de variables
df = df[['DIANOM','MES','TIPOVUELO','OPERA','SIGLADES','temporada_alta','periodo_dia','dif_destino','atraso_15']]
df['MES'] = df['MES'].astype(str)
```

```python
# Generación de variables dummies
df = pd.get_dummies(df, drop_first=True)
# Reorden de columnas
cols = df.columns.tolist()
cols.pop(cols.index('atraso_15'))
cols = cols+['atraso_15']
df.columns = cols
```

```python
# Matriz de correlación
corr_matrix = df.drop(columns='atraso_15').corr().abs()
```

```python
# Selección de triángulo superior
# upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
lower = corr_matrix.where(np.tril(np.ones(corr_matrix.shape), k=-1).astype(np.bool))
```

```python
# Variables con correlación > 0.95
to_drop = [column for column in lower.columns if any(lower[column]>0.95)]
```

```python
# Drop variables
df = df.drop(columns=to_drop)
```

### 1.7 - DEFINICIÓN DE DATA DE ENTRENAMIENTO Y TEST <a class="anchor" id="split"></a>

```python
# Split de data
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,:-1], df.iloc[:,-1], test_size=0.2, random_state=2)
```

### 1.8 - BALANCEO DE DUMMIE 'ATRASO'<a class="anchor" id="bal"></a>

##### Dada la asimetría existente entre la frecuencia de los valores de la variable dependiente "atraso" y "no atraso", se realizará un oversample de la base de datos, utilizando la metodología SMOTE, con el fin de balancear ambos valores.

```python
# Balanceo
oversample = SMOTE()
X_train, y_train = oversample.fit_resample(X_train, y_train)
```

```python
# Inclusión de constante
X_train_cons = sm.add_constant(X_train)
X_test_cons = sm.add_constant(X_test)
```

## 2 - MODELAMIENTO DE DATOS <a class="anchor" id="model"></a>

##### A continuación, se desarrollarán 3 modelos de regresión, con el fin de predecir la variable "atraso". Los modelos fueron escogidos en función del formato de la variable dependiente, al ser un modelo de probabilidad, se descartan otros modelo (por ejemplo: regresión lineal).

### 2.1 MODELO 1: LOGIT <a class="anchor" id="glm"></a>

```python
# Entrenamiento de modelo logit
L_model = LogisticRegression().fit(X_train, y_train)
```

```python
# Predicción de data test
L_y_pred = L_model.predict(X_test)
```

```python
# Performance de modelo logit
print('Accuracy: %.3f' % accuracy_score(y_test, L_y_pred))
print('Precision: %.3f' % precision_score(y_test, L_y_pred))
print('Recall: %.3f' % recall_score(y_test, L_y_pred))
print('F1: %.3f' % f1_score(y_test, L_y_pred))
```

```python
# Plot de coeficientes
L_coeff = pd.DataFrame({'var':L_model.feature_names_in_, 'coefficient':abs(L_model.coef_[0])}).sort_values(by='coefficient')
fig, ax = plt.subplots(figsize=(6,20))
plt.barh(L_coeff['var'], L_coeff['coefficient'])
plt.title('Gráfico de valor absoluto de coeficientes del modelo logit')
plt.grid(axis='x')
ax.set_axisbelow(True)
```

##### Podemos observar que el modelo logit posee un accuracy de $69.2\%$, una precisión de $55.9\%$, un recall de $89.9\%$ y un F1 de $68.9\%$. Entre las variables más importantes consideradas por el modelo se encuentra la variable período_día (tarde), algunos destinos (la serena, punta del este, sao paulo, castro) y algunos operadores (plus ultra).

### MODEL 2: RANDOM FOREST <a class="anchor" id="random"></a>

```python
# Entrenamiento de modelo random forest
RF_model = RandomForestRegressor(n_estimators=20, random_state=2).fit(X_train, y_train) 
```

```python
# Predicción de data test
RF_y_pred = RF_model.predict(X_test)
RF_y_pred = pd.DataFrame([1 if x >= 0.5 else 0 for x in RF_y_pred], index=y_test.index)[0]
```

```python
# Performance de modelo random forest
print('Accuracy: %.3f' % accuracy_score(y_test, RF_y_pred))
print('Precision: %.3f' % precision_score(y_test, RF_y_pred))
print('Recall: %.3f' % recall_score(y_test, RF_y_pred))
print('F1: %.3f' % f1_score(y_test, RF_y_pred))
```

```python
# Plot de features importance
RF_imp = pd.DataFrame({'var':X_train.columns, 'coefficient':RF_model.feature_importances_}).sort_values(by='coefficient')
fig, ax = plt.subplots(figsize=(6,20))
plt.barh(RF_imp['var'], RF_imp['coefficient'])
plt.title('Gráfico de importancia de variables')
plt.grid(axis='x')
ax.set_axisbelow(True)
```

##### Podemos observar que el modelo random forest posee un accuracy de $80.0\%$, una precisión de $58.0\%$, un recall de $76.4\%$ y un F1 de $65.9\%$. Entre las variables más importantes consideradas por el modelo se encuentra la variable período_día (tarde), día de la semana (jueves), temporada_alta,  algunos operadores (united airlines, iberia) y algunos meses (febrero y agosto).

### MODEL 3: NEURAL NETWORK <a class="anchor" id="neural"></a>

```python
# Estructura de modelo random forest
RN_model = Sequential()
# Capa input
RN_model.add(Dense(128, kernel_initializer='normal', input_dim = X_train.shape[1], activation='relu'))
# Capa oculta
RN_model.add(Dense(64, kernel_initializer='normal', activation='relu'))
# Capa output
RN_model.add(Dense(1, kernel_initializer='normal'))
# Compilación de red neuronal
RN_model.compile(loss='mean_absolute_error', optimizer='adam', metrics=['mean_absolute_error'])
```

```python
# Definición de callback
es = EarlyStopping(monitor='val_loss', patience=100, mode='min', restore_best_weights=True, min_delta=1e-10, verbose=2)
```

```python
# Entrenamiento de modelo de red neuronal
RN_model.fit(X_train, y_train, epochs=200, batch_size=32, validation_split=0.2, callbacks=[es])
```

```python
# Predicción de test data
RN_y_pred = RN_model.predict(X_test)
RN_y_pred = pd.DataFrame([1 if x >= 0.5 else 0 for x in RN_y_pred], index=y_test.index)[0]
```

```python
# Performance de modelo red neuronal
print('Accuracy: %.3f' % accuracy_score(y_test, RN_y_pred))
print('Precision: %.3f' % precision_score(y_test, RN_y_pred))
print('Recall: %.3f' % recall_score(y_test, RN_y_pred))
print('F1: %.3f' % f1_score(y_test, RN_y_pred))
```

```python
# SHAP values
background = X_train.sample(n=100)
explainer = shap.KernelExplainer(RN_model.predict, background)
shap_values = explainer.shap_values(shap.sample(X_test, 10))
```

```python
shap.summary_plot(shap_values, X_test, feature_names=X_train.columns)
```

##### Podemos observar que el modelo de redes neuronales posee un accuracy de $74.1\%$, una precisión de $65.8\%$, un recall de $66.0\%$ y un F1 de $65.9\%$. Entre las variables más importantes consideradas por el modelo se encuentra la variable período_día (tarde), algunos operadores (united airlines, aeromexico), algunos destinos (punta cana, miami) y día de la semana (miércoles).

## 3 - CONCLUSIONES <a class="anchor" id="conc"></a>

##### Podemos concluir indiscutiblemente, que la categoría "tarde" de la variable "período día" es la variable más importante, considerada por los 3 modelos para predecir la probabilidad de atraso de un vuelo. Si bien en el análisis descriptivo, no se observa un cambio sustancial en la tasa de atraso para la categoría "tarde", los modelos aislan efectos utilizando un enfoque multivariable.

##### A continuación, podemos ver la tabla con la performance conjunta de los 3 modelos:

| | Logit | Random Forest | Neural Network
| --- | --- | --- | --- |
| Accuracy | 0.692 | 0.700 | 0.741
| Precision | 0.559 | 0.580 | 0.658
| Recall | 0.899 | 0.764 | 0.660
| F1 | 0.689 | 0.659 | 0.659

##### El índice accuracy será descartado como métrica, dada la asimetría de los valores "atraso" y "no atraso", ya que este es un buen indicador sólo en los casos donde la data es balanceada. Por otro lado, el índice de precisión es un buen indicador cuando se busca minimizar los falsos positivos (un predictor de correos SPAM, por ejemplo).

##### En nuestro caso, queremos identificar la mayor cantidad de "atrasos" correctamente, es decir, buscamos minimizar los falsos negativos, y por lo tanto, nos enfocaremos en la métrica recall. Una opción intermedia podría ser F1, pero en este caso en particular, somos indiferentes, ya que ambas métricas se maximizan con el modelo logit.

##### Dicho lo anterior, argumentando la imporancia de identificar los vuelos con atraso, se concluye que el modelo logit es el modelo con mejor performance para nuestro escenario, con un recall de $89.9\%$

##### Como próximos pasos, se recomienda ver la posibilidad de incluir mayor cantidad de datos o mayor cantidad de variables, con el fin de obtener un mejor performance de los modelos predictivos.
